```python
# this is the notebook for regression prediction
# this is the libraries used
import pandas as pd
from scipy.stats import pearsonr
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
# here we load the data
raw_data = pd.read_csv("./Data Set/train.csv")
data = raw_data.copy()
```

```python
data
```

```python
columns = []
for _  in data:
    if type(_) == "str":
        pass
    else:
        columns.extend([_])
        
len(columns)
```

```python
# here we see the shape of the data which we have to predict
# X1 is the input and y1 is the output that is y1 is the target 
#while X1 is the input
X1 = data[['LotFrontage','LotArea','YrSold']]
y1 = data['SalePrice']


# we are going to be starting with linear regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# the libraries are imported and we are using scikit learn to predict the model
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size = 0.2, random_state = 15)
model1 = LinearRegression()
model1.fit(X_train1, y_train1)


# predict on the test set
y_pred1 = model1.predict(X_test1)

# calculate the performance of the model using metrics
mse1 = mean_squared_error(y_test1, y_pred1)
rmse1 = np.sqrt(mse1)
r2_1 = r2_score(y_test1, y_pred1)

# print the performance metrics of the model
print('Mean Squared Error:', mse1)
print('Root Mean Squared Error:', rmse1)
print('R-squared:', r2_1)
```
